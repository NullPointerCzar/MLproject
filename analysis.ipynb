```python
import numpy as np
import re
import matplotlib.pyplot as plt
import pandas as pd
```

<b><h3># Open the chat log file "_chat.txt" for reading with UTF-8 encoding
<h3></b>

```python
f = open("_chat.txt", "r", encoding="utf-8") 
```

<strong><h3>Read the chat log file named "_chat.txt" with UTF-8 encoding and store its content in the variable 'data'.</strong><h3>


```python
data = f.read()
```

```python
print(data)
```

<strong><h3>#Feed the logs instances to AI and let it generate a regex pattern for you
<h3></strong>

```python
pattern = r'^\[(\d{1,2}\/\d{1,2}\/\d{4}),\s*(\d{1,2}:\d{2}(?::\d{2})?)\]\s*(.+?):\s*(.*)$'
```

```python
tuples_list = re.findall(pattern, data, re.MULTILINE)   #pachi pandas dataframe ma convert garna break down to list of tuples
```

```python
tuples_list
```

```python
df = pd.DataFrame(tuples_list, columns=['Date', 'Time', 'Name', 'Message'])
df.head()
```

```python
df['Date']  = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time']  = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
```

```python
df
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day_name()
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
df['Minute'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.minute
```
